## 1. Regressão Linear com Scikit-learn (Validação Cruzada de 5 Pastas)

```python
from sklearn.model_selection import KFold

# Inicializa KFold para 5 pastas
kf = KFold(n_splits=5, shuffle=True, random_state=42)

mse_sklearn_folds = []
pearson_sklearn_folds = []

# Loop para cada pasta da validação cruzada
for fold, (train_index, test_index) in enumerate(kf.split(x)):
    x_train_fold, x_test_fold = x[train_index], x[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]

    # Inicializa e treina o modelo de Regressão Linear
    model_sklearn = LinearRegression()
    model_sklearn.fit(x_train_fold, y_train_fold)

    # Faz as predições
    y_pred_sklearn = model_sklearn.predict(x_test_fold)

    # Calcula o Erro Médio Quadrático (MSE)
    mse = metrics.mean_squared_error(y_test_fold, y_pred_sklearn)
    mse_sklearn_folds.append(mse)

    # Calcula o Coeficiente de Pearson
    # numpy.corrcoef espera arrays 1D, então achatamos se necessário
    if y_pred_sklearn.ndim > 1:
        y_pred_sklearn = y_pred_sklearn.ravel()
    if y_test_fold.ndim > 1:
        y_test_fold = y_test_fold.ravel()

    pearson_corr = np.corrcoef(y_test_fold, y_pred_sklearn)[0, 1]
    pearson_sklearn_folds.append(pearson_corr)

    print(f"--- Pasta {fold + 1} ---")
    print(f"MSE (Scikit-learn): {mse:.4f}")
    print(f"Coeficiente de Pearson (Scikit-learn): {pearson_corr:.4f}")

# Calcula a média dos resultados
mean_mse_sklearn = np.mean(mse_sklearn_folds)
mean_pearson_sklearn = np.mean(pearson_sklearn_folds)

print("\n--- Resultados Médios (Scikit-learn) ---")
print(f"MSE Médio (Scikit-learn): {mean_mse_sklearn:.4f}")
print(f"Coeficiente de Pearson Médio (Scikit-learn): {mean_pearson_sklearn:.4f}")
```

## 2. Regressão Linear com Pseudo-Inversa (Validação Cruzada de 5 Pastas)

```python
mse_pseudo_inverse_folds = []
pearson_pseudo_inverse_folds = []

# Loop para cada pasta da validação cruzada
for fold, (train_index, test_index) in enumerate(kf.split(x)):
    x_train_fold, x_test_fold = x[train_index], x[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]

    # Adiciona a coluna de bias (1s) para o termo independente
    bias_train = np.ones((x_train_fold.shape[0], 1))
    x_train_fold_biased = np.hstack((x_train_fold, bias_train))

    bias_test = np.ones((x_test_fold.shape[0], 1))
    x_test_fold_biased = np.hstack((x_test_fold, bias_test))

    # Calcula os pesos (w) usando a pseudo-inversa
    # w = (X_transposed @ X)^-1 @ X_transposed @ Y
    w = np.linalg.inv(x_train_fold_biased.T @ x_train_fold_biased) @ x_train_fold_biased.T @ y_train_fold

    # Faz as predições
    y_pred_pseudo_inverse = x_test_fold_biased @ w

    # Calcula o Erro Médio Quadrático (MSE)
    mse = metrics.mean_squared_error(y_test_fold, y_pred_pseudo_inverse)
    mse_pseudo_inverse_folds.append(mse)

    # Calcula o Coeficiente de Pearson
    # numpy.corrcoef espera arrays 1D, então achatamos se necessário
    if y_pred_pseudo_inverse.ndim > 1:
        y_pred_pseudo_inverse = y_pred_pseudo_inverse.ravel()
    if y_test_fold.ndim > 1:
        y_test_fold = y_test_fold.ravel()

    pearson_corr = np.corrcoef(y_test_fold, y_pred_pseudo_inverse)[0, 1]
    pearson_pseudo_inverse_folds.append(pearson_corr)

    print(f"--- Pasta {fold + 1} ---")
    print(f"MSE (Pseudo-Inversa): {mse:.4f}")
    print(f"Coeficiente de Pearson (Pseudo-Inversa): {pearson_corr:.4f}")

# Calcula a média dos resultados
mean_mse_pseudo_inverse = np.mean(mse_pseudo_inverse_folds)
mean_pearson_pseudo_inverse = np.mean(pearson_pseudo_inverse_folds)

print("\n--- Resultados Médios (Pseudo-Inversa) ---")
print(f"MSE Médio (Pseudo-Inversa): {mean_mse_pseudo_inverse:.4f}")
print(f"Coeficiente de Pearson Médio (Pseudo-Inversa): {mean_pearson_pseudo_inverse:.4f}")
```

## 3. Comparação dos Resultados

```python
print("--- Comparação dos Métodos ---")
print(f"Scikit-learn - MSE Médio: {mean_mse_sklearn:.4f} | Pearson Médio: {mean_pearson_sklearn:.4f}")
print(f"Pseudo-Inversa - MSE Médio: {mean_mse_pseudo_inverse:.4f} | Pearson Médio: {mean_pearson_pseudo_inverse:.4f}")

if mean_mse_sklearn < mean_mse_pseudo_inverse:
    print("\nO modelo Scikit-learn apresentou um MSE médio menor, indicando um desempenho ligeiramente melhor.")
elif mean_mse_pseudo_inverse < mean_mse_sklearn:
    print("\nO modelo da Pseudo-Inversa apresentou um MSE médio menor, indicando um desempenho ligeiramente melhor.")
else:
    print("\nAmbos os modelos apresentaram MSEs médios semelhantes.")

if mean_pearson_sklearn > mean_pearson_pseudo_inverse:
    print("O modelo Scikit-learn apresentou um Coeficiente de Pearson médio maior, indicando uma correlação ligeiramente mais forte.")
elif mean_pearson_pseudo_inverse > mean_pearson_sklearn:
    print("O modelo da Pseudo-Inversa apresentou um Coeficiente de Pearson médio maior, indicando uma correlação ligeiramente mais forte.")
else:
    print("Ambos os modelos apresentaram Coeficientes de Pearson médios semelhantes.")
```
